```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy
import pandas
import tensorflow as tf
import time
import datetime
import os
```

```python
# Historical data retrieval, HK Stock

root = os.getcwd()

stock_id = "0005.hk"
market = "hk"
stock_ticket = yf.Ticker(stock_id) 
path = 'stock_history/' + market + '/'

try:
    os.system('mkdir ' + path)
    os.system('mkdir ' + path + stock_id)
except:
    print('file_exist')
    
    
for i in range(-29,1):
    start_time = str(datetime.date.today() + datetime.timedelta(days=i))
    end_time = str(datetime.date.today() + datetime.timedelta(days=i+1))
    print('start: '+ start_time)
    print('end: '+ end_time)
#     try:
    price_historical = stock_ticket.history(start=start_time, end=end_time, interval="1m")
    file_name = start_time + '.csv'
    price_historical.to_csv(path + stock_id + '/' + file_name)
    print(price_historical)
#     except:
#         print('some error, Historical data retrieval, HK Stock')
```

```python
# Day New data retrieval, HK Stock

root = os.getcwd()

stock_id = "0005.hk"
market = "hk"
stock_ticket = yf.Ticker(stock_id) 
path = 'stock_history/' + market + '/'

try:
    os.system('mkdir ' + path)
    os.system('mkdir ' + path + stock_id)
except:
    print('file_exist')
    
while True:
    for i in range(0,1):
        start_time = str(datetime.date.today() + datetime.timedelta(days=i))
        end_time = str(datetime.date.today() + datetime.timedelta(days=i+1))
        print('start: '+ start_time)
#         print('end: '+ end_time)
    #     try:
        price_historical = stock_ticket.history(start=start_time, end=end_time, interval="1m")
        file_name = start_time + '.csv'
        price_historical.to_csv(path + stock_id + '/' + file_name)
#         print(price_historical)
    #     except:
    #         print('some error, Historical data retrieval, HK Stock')
    time.sleep(60)
```
